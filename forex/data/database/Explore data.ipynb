```python
def fill0(num):
    s = str(num)
    return "0" * (2 - len(s)) + s
```

```python
def transform(file):
    file_split = file.split("_")
    pair = file_split[2]
    df = pd.read_excel(file, sheet_name=0, header=None)
    
    data = OrderedDict()
    for index, row in df.iterrows():
        t = row[0]
        key = t.strftime("%Y-%m-%d-%H") + "-" + fill0(15*(t.minute // 15))
        if key not in data:
            data[key] = []
        data[key].append(row[1])
    
    data2 = {}
    for k, vs in data.items():
        vs = np.asarray(vs)
        #open-high-low-close
        data2[k] = [vs[0], vs.max(), vs.min(), vs[-1]]
    
    return pair, data2
```

```python
def load(con, pair, data):
    sql_stmt = "INSERT INTO FOREX_SPLIT (timestmp, pair, O, H, L, C) VALUES (?, ?, ?, ?, ?, ?)"
    data_list = [(datetime.datetime.strptime(k, "%Y-%m-%d-%H-%M").timestamp(), pair, v[0], v[1], v[2], v[3]) for k, v in data.items()]
    con.executemany(sql_stmt, data_list)
```

```python
con = sqlite3.connect("FOREX")
with open("schema.sql") as f:
    content = "".join(f.readlines())
    con.executescript(content)
```

```python
PATH = "..\\files\\data"
dir = sorted(os.listdir(PATH))
for file in list(filter(lambda a: "xls" in a, dir)):
    file_path = PATH + "\\" + file
    print(file)
    pair, data = transform(file_path)
    print("transformed")
    load(con, pair, data)
    print("loaded")
```

```python
def find_sections(list):
    sections = []
    curr_section = []
    previous = None
    delta = datetime.timedelta(minutes=15)
    in_section = False
    for v in list:
        dt_1 = datetime.datetime.fromtimestamp(v[0])
        
        if previous != None:
            dt_0 = datetime.datetime.fromtimestamp(previous[0])

            if dt_0 + delta == dt_1:
                if in_section:
                    #it continues here
                    curr_section.append(v)
                else:
                    #it starts here
                    curr_section = [previous, v]
                    in_section = True
            else:
                if in_section:
                    #it ends here
                    sections.append(curr_section)
                    in_section = False

        previous = v 
        
    return sections
```
